```python
X_train=[]
y_train=[]
X_test=[]
y_test=[]
for l in open('kolos.syn.train'):
    l=l.split('\t')
    X_train.append((l[2]+'#'+l[3],l[5]+'#'+l[6]))
    y_train.append(l[7])
for l in open('kolos.syn.test'):
    l=l.split('\t')
    X_test.append((l[2]+'#'+l[3],l[5]+'#'+l[6]))
    y_test.append(l[7])
lexicon=set([e[0] for e in X_test]+[e[1] for e in X_test]+[e[0] for e in X_train]+[e[1] for e in X_train])
print len(lexicon)
```

```python
import numpy as np
embeddings={}
for l in open('embed.sl-lemma.ft.sg.vec'):
    if len(l)>30:
        l=l.strip().split(' ')
        if l[0] in lexicon:
            embeddings[l[0]]=np.array([float(e) for e in l[1:]])
print len(embeddings)
```

```python
X_train_emb=[]
y_train_emb=[]
lex_train_emb=[]
X_test_emb=[]
y_test_emb=[]
lex_test_emb=[]
for X,y in zip(X_train,y_train):
    if X[0] in embeddings and X[1] in embeddings:
        X_train_emb.append(np.concatenate((embeddings[X[0]],embeddings[X[1]])))
        X_train_emb.append(np.concatenate((embeddings[X[1]],embeddings[X[0]])))
        y_train_emb.append(int(y))
        y_train_emb.append(int(y))
        lex_train_emb.append((X[0],X[1]))
        lex_train_emb.append((X[1],X[0]))
print len(X_train_emb)
for X,y in zip(X_test,y_test):
    if X[0] in embeddings and X[1] in embeddings:
        X_test_emb.append(np.concatenate((embeddings[X[0]],embeddings[X[1]])))
        X_test_emb.append(np.concatenate((embeddings[X[1]],embeddings[X[0]])))
        y_test_emb.append(int(y))
        y_test_emb.append(int(y))
        lex_test_emb.append((X[0],X[1]))
        lex_test_emb.append((X[1],X[0]))
print len(X_test_emb)
print Counter([e[0][-2:] for e in lex_train_emb])
from keras.utils import to_categorical
y_train_emb=to_categorical(y_train_emb)
y_test_emb=to_categorical(y_test_emb)
```

```python
X_train_emb_pos={'Nc':[],'Vm':[],'Ag':[],'all':[]}
y_train_emb_pos={'Nc':[],'Vm':[],'Ag':[],'all':[]}
lex_train_emb_pos={'Nc':[],'Vm':[],'Ag':[],'all':[]}
for l,x,y in zip(lex_train_emb,X_train_emb,y_train_emb):
    pos=l[0][-2:]
    if pos in X_train_emb_pos:
        X_train_emb_pos[pos].append(x)
        y_train_emb_pos[pos].append(y)
        lex_train_emb_pos[pos].append(l)
        X_train_emb_pos['all'].append(x)
        y_train_emb_pos['all'].append(y)
        lex_train_emb_pos['all'].append(l)        
X_test_emb_pos={'Nc':[],'Vm':[],'Ag':[],'all':[]}
y_test_emb_pos={'Nc':[],'Vm':[],'Ag':[],'all':[]}
lex_test_emb_pos={'Nc':[],'Vm':[],'Ag':[],'all':[]}
for l,x,y in zip(lex_test_emb,X_test_emb,y_test_emb):
    pos=l[0][-2:]
    if pos in X_train_emb_pos:
        X_test_emb_pos[pos].append(x)
        y_test_emb_pos[pos].append(y)
        lex_test_emb_pos[pos].append(l)
        X_test_emb_pos['all'].append(x)
        y_test_emb_pos['all'].append(y)
        lex_test_emb_pos['all'].append(l)        
```

```python
pos='all'
from keras.models import Sequential, Model
from keras.layers import Dense, Activation, Dropout, Concatenate, Input
import tensorflow as tf
from sklearn.metrics import roc_auc_score
from keras.optimizers import RMSprop
def auc_roc(y_true, y_pred):    
    return tf.py_func(roc_auc_score, (y_true, y_pred), tf.double)
#y_train_emb_cat = to_categorical(y_train_emb, num_classes=2)
from collections import Counter
#print Counter(y_train_emb)
large_input=Input(shape=(200,))
intermediate=Dense(64,activation='relu')(large_input)
dropout=Dropout(0.5)(intermediate)
out=Dense(2,activation='softmax')(dropout)
model=Model(inputs=large_input,outputs=out)
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['acc',auc_roc])
model.fit(np.array(X_train_emb_pos[pos]), np.array(y_train_emb_pos[pos]), epochs=20, batch_size=512, class_weight={0:3,1:1})
```

```python
y_test_prob=model.predict(np.array(X_test_emb_pos[pos]))
y_test_pred=np.argmax(y_test_prob,axis=1)
from sklearn.metrics import confusion_matrix,roc_auc_score
print roc_auc_score([e[0] for e in y_test_emb_pos[pos]],[e[0] for e in y_test_prob])
for p in ('Nc','Vm','Ag'):
    y_test_pred_p=model.predict(np.array(X_test_emb_pos[p]))
    print p, roc_auc_score([e[0] for e in y_test_emb_pos[p]],[e[0] for e in y_test_pred_p])

    
probs=sorted(zip([e[1] for e in y_test_prob],[int(e[1]) for e in y_test_emb_pos[pos]],lex_test_emb_pos[pos]))
for prob in probs:
    print prob[0],prob[1],prob[2][0],prob[2][1]
```

```python
def pair(X):
    new_X=[list() for i in range(100)]
    for x in X:
        for i in range(100):
            new_X[i].append((x[i],x[i+100]))
    return [np.array(e) for e in new_X]
inputs=[]
denses=[]
for i in range(100):
    single_input=Input(shape=(2,))
    dense=Dense(2,activation='relu')(single_input)
    inputs.append(single_input)
    denses.append(dense)
merged=Concatenate()(denses)
dropout=Dropout(0.5)(merged)
out=Dense(2,activation='softmax')(dropout)
hmodel=Model(inputs=inputs,outputs=out)
hmodel.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['acc',auc_roc])
hmodel.fit(pair(X_train_emb_pos[pos]), np.array(y_train_emb_pos[pos]), epochs=20, batch_size=512,class_weight={0:3,1:1})
```

```python
y_test_prob=hmodel.predict(pair(X_test_emb_pos[pos]))
y_test_pred=np.argmax(y_test_prob,axis=1)
from sklearn.metrics import confusion_matrix,roc_auc_score
print roc_auc_score([e[0] for e in y_test_emb_pos[pos]],[e[0] for e in y_test_prob])
probs=sorted(zip([e[1] for e in y_test_prob],[int(e[1]) for e in y_test_emb_pos[pos]],lex_test_emb_pos[pos]))
for prob in probs:
    print prob[0],prob[1],prob[2][0],prob[2][1]
```
